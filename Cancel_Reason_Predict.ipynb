```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import ExcelWriter
from pandas import ExcelFile

#load the train dataset
df = pd.read_excel('Junior Data Analyst Data for Case Study.xlsx', engine='openpyxl', sheet_name='Query result')
df.head()
```

```python
df['id'] = range(1, len(df) + 1)
```

```python
df.info()
```

```python
df= df.fillna(0)
```

```python
df.info()
```

```python
df['Cancellation Time'] = pd.to_timedelta(df['Cancellation Time'].astype(str))

df['diff'] = df['Cancellation Time'].diff()
df['mask'] = df['Cancellation Time'].diff() > pd.Timedelta(5, unit='s')
df.head()
```

```python
df['Cancellation Time'] = pd.to_timedelta(df['Cancellation Time'])
diff = df['Cancellation Time'].diff()
mask = df['Cancellation Time'].diff() > pd.Timedelta(5, unit='s')
df['new'] = diff.where(mask)
df.head()
```

```python
df[['hour','minute','second']] = pd.to_timedelta(df['Cancellation Time']).dt.components.iloc[:, 1:4]
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

labelEn = LabelEncoder()

df['Personal or Business'] = labelEn.fit_transform(df['Personal or Business'].astype(str))
df['Vehicle Type'] = labelEn.fit_transform(df['Vehicle Type'].astype(str))
df['Standard or Express'] = labelEn.fit_transform(df['Standard or Express'].astype(str))
df['Platform'] = labelEn.fit_transform(df['Platform'].astype(str))
df['Cancel Reason'] = labelEn.fit_transform(df['Cancel Reason'].astype(str))
df['Cancelled by'] = labelEn.fit_transform(df['Cancelled by'].astype(str))
df['scheduled'] = labelEn.fit_transform(df['scheduled'].astype(str))
```

```python
#dropping the columns we don't need
df = df.drop(['Placement Time','Cancellation Time','Confirmation Time','diff','mask','new'], axis = 1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split, KFold, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import precision_score, recall_score, precision_recall_curve,f1_score, fbeta_score, confusion_matrix, make_scorer, accuracy_score
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, auc, log_loss
```

```python
X1, y1 = df.drop(columns=['Cancel Reason','id'],axis=1),df['Cancel Reason']

from imblearn.over_sampling import SMOTE
X, y = SMOTE(random_state=42).fit_resample(X1,y1)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2)
```

```python
forest = RandomForestClassifier(n_estimators=100, max_depth=7, max_features=12)
forest.fit(X_train, y_train)
forest.score(X_train, y_train)

print("The score for Random Forest is")
#measuring the accuracy of the model against the training data(How well the model explains the data it was trained with)
print("Training: {:6.2f}%".format(100*forest.score(X_train, y_train)))
```

```python
forest_pred = forest.predict(X_test)

accuracy = accuracy_score(y_test, forest_pred)
print("Accuracy: %.3f%%" % (accuracy * 100.0))
```

```python

from lightgbm import LGBMClassifier
light = LGBMClassifier()
light.fit(X_train, y_train)
lightP = light.predict(X_test)

lightAS = accuracy_score(y_test, lightP)
lightaccuracy = lightAS*100
lightaccuracy
```

```python
print(classification_report(y_test, lightP))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, y_train)
knnP = knn.predict(X_test)

knnAS = accuracy_score(y_test, knnP)
knnaccuracy = knnAS*100
knnaccuracy
```

```python
#Using Boosting model

from sklearn.ensemble import GradientBoostingClassifier

boostmodel = GradientBoostingClassifier()
boostmodel.fit(X_train, y_train)
boostmodelP = boostmodel.predict(X_test)

boostmodelAS = accuracy_score(y_test, boostmodelP)
boostaccuracy = boostmodelAS*100
boostaccuracy
```

```python
#Use the new model to predict a new set of data
test_predictions = light.predict(X_test)
print(test_predictions)
```
